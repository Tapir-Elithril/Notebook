### TAG0: Example,  Read an image using imageio and opencv.

Hint: Read the documentation of opencv and imageio on imread.


```python
def read_image_cv_io(img_path):
    '''
    Input:
    img_path: string
    
    Return:
    img_cv: array, (H, W, 3)
    img_io: array, (H, W, 3)
    '''
    ##################################
    img_io = imageio.imread(img_path) #RGB
    img_cv = cv2.imread(img_path) #BGR
    return img_cv, img_io
    ##################################

img_path = 'lena.png'
img_cv, img_io = read_image_cv_io(img_path)
print('imageio image shape: \n', img_io.shape, '\nimage: \n', img_io)
print('opencv image shape: \n' , img_cv.shape, '\nimage: \n', img_cv)
```

```python
def concat_images(images):
    '''
    Input: 
    images: tuple of two images, ((H, W, 3), (H, W, 3))
    Output:
    image: (H, 2*W, 3)
    '''
    ########################
    # Write your code here
    result = np.concatenate(images, axis=1) # concatenate along width
    return result
    ########################
    
show_image(concat_images((img_io, img_cv)))
```

## OpenCV

### TAG5: Convert and show images.

Note that you need to add a waitKey function after calling imshow in opencv.

Remember to destroy the windows using cv2.destroyAllWindows.

```python
def convert_rgb_to_bgr(img):
    '''
    Input:
    img: (H, W, 3), RGB channels.
    Return:
    img: (H, W, 3), BGR channels.
    '''
    ########################
    # Write your code here
    bgr_img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
    return bgr_img
    ########################

def convert_rgb_to_gray(img):
    '''
    Input:
    img: (H, W, 3), RGB channels.
    Return:
    img: (H, W), Gray image.
    '''
    ########################
    # Write your code here
    gray_img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    return gray_img
    ########################

def cv_imshow(img):
    '''
    Input:
    img: (H, W, 3), BGR channels
    '''
    ########################
    # Write your code here
    window_name = 'Image'
    cv2.imshow(window_name,img) # BGR channel show
    cv2.waitKey(0) # forever waiting,until press key
    ########################
    
cv_imshow(convert_rgb_to_bgr(img_io)) # red -> blue -> red
cv_imshow(convert_rgb_to_gray(img_io))
cv2.destroyAllWindows()
```

### TAG6: Read videos and show videos.

Try to wait 10 ms between frames.

```python
def read_and_show_video(video_path):
    '''
    Input: 
    video_path: string
    '''
    ########################
    # Write your code here
    cap = cv2.VideoCapture(video_path) # VideoCapture creates video object
    while True:
        ret, frame = cap.read() # ret:is finished, frame = image
        if not ret:
            break

        cv2.imshow("Video", frame)
        key = cv2.waitKey(10) & 0xFF # wait 10 ms between frames
        # return 0 if nothing happens,else return the ASCII code of the keyboard input
        if key == ord('q') or key == 27: # press q or ESC
            break
    cap.release() # release the video capture object after displayed
    ########################

read_and_show_video('sfm.mp4')
cv2.destroyAllWindows()
```
